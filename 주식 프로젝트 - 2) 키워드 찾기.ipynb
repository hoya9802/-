# 1. 뉴스 제목 데이터 가져오기

```python
chrome_loc = 'C:/pywork/chromedriver.exe'
url_target = 'https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2=259&sid1=101&date=20210723'
driver = webdriver.Chrome(chrome_loc)
driver.get(url_target)
```

```python
news_dates = driver.find_element_by_class_name('pagenavi_day')
news_dates_text = news_dates.text
print(news_dates_text)
```

```python
last_date = 701
current_date = news_dates_text[0:7]
current= re.findall('[1-99]',current_date)
if len(current[1]) == 1:
    current = current[0]+str('0')+str(current[1])
```

## 1) 금융 카테고리 뉴스 제목 가져오기

```python
finance_text = []
words_list = []

for i in np.arange(723, last_date, -1):
    if int(i/100) < 10:
        temp_date =  '0'+ str(i)
    else:
        temp_date = str(i)
    finance_date_link = 'https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2=259&sid1=101&date=2021'+temp_date
    driver.get(finance_date_link)    
    print('{0}월 {1}일 페이지 접속'.format(int(i/100), i%100))
    
    pages = driver.find_element_by_class_name('paging').text
    page_list = re.findall('\d+|\w+', pages)
    
    while page_list[-1] == '다음':
        button = driver.find_element_by_css_selector('#main_content > div.paging > a.next.nclicks\(fls\.page\)')
        button.click()
        x = driver.find_element_by_class_name('paging')
        y = x.text
        w = re.findall('\d+|\w+', y)
        page_list += w
    page_num_list = [i for i in page_list if i != '이전' and i != '다음']
    del page_num_list[-1]
    print('총 페이지 수:', page_num_list)
    print('===============================================================================')
    time.sleep(.2)

    for j in page_num_list:
        xpath = finance_date_link+'&page='+ j
        driver.get(xpath)
        for k in [1, 2]:
            for w in range(1, 11):
                news_title = driver.find_element_by_xpath('//*[@id="main_content"]/div[2]/ul['+str(k)+']/li['+str(w)+']/dl')
                data_text = news_title.text
                finance_text.append(data_text)
                time.sleep(0.2)
                print('{0}의 {1} 기사 불러오기 완료'.format(k, w))
        print('{}page 완료'.format(j))
        time.sleep(.2)
```

```python
#데이터프레임 만들기
finance_text_list = list()

for i in range(0, len(finance_text)):
    temp = finance_text[i].split('\n')
    finance_text_list += [temp[0]]
    
finance_text_df = pd.DataFrame(finance_text_list)

#데이터프레임을 csv로 저장하기
finance_text_df.to_csv('c:/pywork/finance_text.csv', encoding = 'utf-8')
```
